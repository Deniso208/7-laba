```python
import random
import matplotlib.pyplot as plt

def read_array_from_file(filename):
    with open(filename, 'r') as file:
        array = [int(line) for line in file.readlines()]
    return array

def visualize_array(array):
    min_value = min(array)
    max_value = max(array)
    
    colors = ['g' if x > 0 else 'b' if x == 0 else 'r' for x in array]

    plt.bar(range(len(array)), array, color=colors)
    
 # Замініть color='g' на color='r' і навпаки для відображення правильно
    plt.axhline(y=min_value, color='r', linestyle='--', label=f'Max: {min_value}')
    plt.axhline(y=max_value, color='g', linestyle='--', label=f'Min: {max_value}')

    plt.legend()
    plt.xlabel('Індекс елементу')
    plt.ylabel('Значення')
    plt.title('Візуалізація масиву')
    plt.show()

def sum_positive_elements(array):
    positive_elements = [x for x in array if x > 0]
    return sum(positive_elements)

def count_negative_elements_between_zeros(array):
    first_zero_index = array.index(0)
    last_zero_index = len(array) - array[::-1].index(0) - 1
    negative_elements_between_zeros = [x for x in array[first_zero_index+1:last_zero_index] if x < 0]
    return len(negative_elements_between_zeros)

def transform_array(array):
    positive_elements = sorted([x for x in array if x > 0])
    zero_elements = [0] * array.count(0)
    negative_elements = sorted([x for x in array if x < 0], reverse=True)
    transformed_array = positive_elements + zero_elements + negative_elements
    return transformed_array

if __name__ == '__main__':
    filename = 'generated_array.txt'
    generated_array = read_array_from_file(filename)
    
    print("Зчитаний масив:", generated_array)
    visualize_array(generated_array)
    
    sum_positive = sum_positive_elements(generated_array)
    print("Сума додатних елементів:", sum_positive)
    
    count_negatives_between_zeros = count_negative_elements_between_zeros(generated_array)
    print("Кількість від'ємних елементів між нулями:", count_negatives_between_zeros)
    
    transformed_array = transform_array(generated_array)
    print("Масив після перетворення:", transformed_array)
    visualize_array(transformed_array)
```
